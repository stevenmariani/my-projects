## Data Analysis (T-test, ANOVA, Regression)  

### Task 1:  is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)?

<b> Hypotesis </b> <br>
H0: there isn't a significant difference in MEDV between areas bounded by the Charles River <br>
H1: there is a significant difference in MEDV between areas bounded by the Charles River

<b> Alpha </b> = 0.05

```python
#Variance validation with Levene test

alpha = 0.05

stat, p = scipy.stats.levene(boston_df[boston_df['CHAS'] == 0.00]['MEDV'],
                   boston_df[boston_df['CHAS'] == 1.00]['MEDV'], center='mean')

print(f"Results: Statistics = {stat:.3f}, p-value = {p:.3g}")

if p < alpha:
    print("Since p < alpha we reject the null hypotesis and we can assume there are different variances")
else:
    print("Since p > alpha we accept the null hypotesis and we can assume there are equal variances")
```

```python
#Since the p-value of Levene test is less than 0.05 I will assume different variances

t_stat, p_value = scipy.stats.ttest_ind(boston_df[boston_df['CHAS'] == 0.00]['MEDV'],
                   boston_df[boston_df['CHAS'] == 1.00]['MEDV'], equal_var = False)

print(f"Results: T-statistic= {t_stat}, p-value: {p_value}")

if p_value < alpha:
    print("Since p_value < alpha we reject the null hypotesis and can affirm that there is a significant difference\
      \nbetween the median value of houses bounded by the Charles River and the houses not bounded.")
else:
    print("Since p_value > alpha we accept the null hypotesis and we can affirm that there isn't a significant difference\
    \n between the median value of houses bounded by the Charles River and the houses not bounded.")
```

### Task 2:  is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)

<b> Hypotesis </b> <br>
H0: there isn't a significant difference in MEDV for each proportion of owner occupied units built prior to 1940 <br>
H1: there is a significant difference in MEDV for each proportion of owner occupied units built prior to 1940

<b> Alpha </b> = 0.05

```python
#Levene test to validate equal variances

stat, p = scipy.stats.levene(boston_df[boston_df['age_group'] == '35 years and younger']['MEDV'],
                   boston_df[boston_df['age_group'] == 'between 35 and 70 years']['MEDV'], 
                   boston_df[boston_df['age_group'] == '70 years and older']['MEDV'], 
                   center='mean')

print(f"Results: Statistics = {stat:.3f}, p-value = {p:.3g}")

if p < alpha:
    print("Since p < alpha we reject the null hypotesis and we can assume there are different variances")
else:
    print("Since p > alpha we accept the null hypotesis and we can assume there are equal variances")
```

```python
#Since the p-value of the Levene test is greater that alpha we assume equal variance

less_than_35 = boston_df[boston_df['age_group'] == '35 years and younger']['MEDV']
from_35_to_70 = boston_df[boston_df['age_group'] == 'between 35 and 70 years']['MEDV'] 
greater_than_70 = boston_df[boston_df['age_group'] == '70 years and older']['MEDV']

f_statistic, p_value = scipy.stats.f_oneway(less_than_35, from_35_to_70, greater_than_70)

print(f"Results: F_Statistic = {f_statistic:.3f}, p-value = {p:.3g}")

if p_value < alpha:
    print("Since p_value < alpha we reject the null hypotesis and can affirm that there is significant differences in median value\
      \nbetween each proportion of owner occupied units built prior to 1940.")
else:
    print("Since p_value > alpha we accept the null hypotesis and we can affirm that there isn't significant differences in median value\
    \nbetween each proportion of owner occupied units built prior to 1940.")
```

### Task 3: Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation) 

<b> Hypotesis </b> <br>
H0: there isn't a significant correlation between nitric oxide concentrations and proportion of non-retail business acres per town <br>
H1: there is a significant correlation between nitric oxide concentrations and proportion of non-retail business acres per town

<b> Alpha </b> = 0.05

```python
corr_coefficient, p_value = scipy.stats.pearsonr(boston_df["NOX"], boston_df["INDUS"])

print(f"Results: R = {corr_coefficient:.3f}, p-value = {p:.3g}")

if p_value < alpha:
    print("Since p_value < alpha we reject the null hypotesis and can affirm that there is a significant correlation between\
    \nitric oxide concentrations and proportion of non-retail business acres per town.")
else:
    print("Since p_value > alpha we accept the null hypotesis and can affirm that there isn't a significant correlation between\
    \nitric oxide concentrations and proportion of non-retail business acres per town.")
```
